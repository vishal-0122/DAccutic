```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importing the csv
df = pd.read_csv('adult.csv')
df.head()
```

```python
df.shape
```

```python
# Renaming the attributes
df = df.set_axis(['Age','Workclass','FNLWGT','Education','Edu_No.','Marital_Status','Occupation','Relationship','Race','Sex','Cptl_Gain','Cptl_Loss','Hours_p/w','Country','Income'],axis = 1)
df.head()
```

```python
df.drop(['Marital_Status','Relationship','Race','Sex','Country'],axis=1,inplace=True)
df.head()
```

```python
# Encoding the categorical values
from sklearn.preprocessing import LabelEncoder

le_workclass = LabelEncoder()
le_education = LabelEncoder()
le_occupation = LabelEncoder()
le_income = LabelEncoder()

df['Workclass_n'] = le_workclass.fit_transform(df['Workclass'])
df['Education_n'] = le_education.fit_transform(df['Education'])
df['Occupation_n'] = le_occupation.fit_transform(df['Occupation'])
df['Income_n'] = le_income.fit_transform(df['Income'])
df.drop(['Workclass','Education','Occupation','Income'], axis='columns',inplace=True)
df
```

```python
x = df.iloc[:,:-1] #ind
y = df.iloc[:,-1]  #dep
print(type(x))
print(type(y))
print(x.shape)
print(y.shape)
```

```python
#splitting the dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# 1 --> Applying Decision Tree Model
from sklearn.tree import DecisionTreeClassifier
```

```python
mod1 = DecisionTreeClassifier(criterion='gini')
mod1.fit(x_train,y_train)
```

```python
#Accuracy 
print("Training_Score",mod1.score(x_train,y_train))
print("Testing_Score",mod1.score(x_test,y_test))
```

```python
# Computing the % of misclassification
mc_m1= 1 - mod1.score(x_test,y_test)
print('Misclassification Error = ',mc_m1* 100,'%')
```

```python
# Testing the Decision Tree
ypred_m1 = mod1.predict(x_test)
print(ypred_m1)
```

```python
# Computing Confusion Matrix & Classification Report
from sklearn.metrics import confusion_matrix, classification_report
c_m1 = confusion_matrix(y_test,ypred_m1)
print(c_m1)
print(classification_report(y_test,ypred_m1))
```

```python
# 2. --> Applying Random Forest Classifer
from sklearn.ensemble import RandomForestClassifier

mod2 = RandomForestClassifier(n_estimators=60,criterion='entropy')
mod2.fit(x_train,y_train)
```

```python
#Accuracy 
print("Training_Score",mod2.score(x_train,y_train))
print("Testing_Score",mod2.score(x_test,y_test))
```

```python
# Testing Random Forest
ypred_m2 = mod2.predict(x_test)
print(ypred_m2)
```
